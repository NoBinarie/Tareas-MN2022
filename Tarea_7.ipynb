```python
import matplotlib.pyplot as plt #importamos las librerias necesarias
import numpy as np
```

```python
def gauss(x): #definimos una funcion de distribucion gaussiana, con los parametros mu y sigma constantes.
    return np.exp(-(x)**2/(2*1))/(np.sqrt(2*np.pi))
```

```python
def lagrange(x, puntos): #definimos la funcion de polinomios de lagrange   
    Lis = np.ones(len(puntos))
    fs = np.ones(len(puntos))
    for i in range(len(puntos)):
        fs[i] = puntos[i][1]
        for j in range(len(puntos)):
            if j != i:
                Lis[i] *= (x - puntos[j][0])/(puntos[i][0] - puntos[j][0])    
    return sum(fs*Lis)
```

```python
x=np.linspace(0,2,21) #creamos el intervalo x=[0,2]
plt.plot(x,gauss(x)) #graficamos la gaussiana de el intervalo
```

```python
puntos= np.stack((x,gauss(x)),axis=1) #creamos el array de puntos x,y de la gaussiana
```

```python
plt.plot(x,gauss(x))  #ahora graficamos los puntos de la interpolacion junto a el plot de la gaussiana, podemos ver que es una
plt.scatter(x,[lagrange(xi, puntos) for xi in x]) #buena aproximacion
```

```python
xn=np.linspace(0,3,21) #ahora graficamos la gaussiana para intervalo x=[0,3] y usamos la interpolacion anterior, podemos ver
plt.plot(xn,gauss(xn)) #que la aproximacion no sirve, pues dados 𝑛+1  puntos, hay un único y solo un unico polinomio que pasa
plt.scatter(xn,[lagrange(xi, puntos) for xi in x]) # por esos puntos, al agregar el 3 deberiamos cambiar el polinomio.
```

```python
plt.scatter(xn,[lagrange(xi, puntos) for xi in xn]) #asi
plt.plot(xn,gauss(xn))
```
